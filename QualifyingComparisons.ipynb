# We can use Qualifying data to examine speed differences between team-mates/cars

## The QualyExtractor class inherits from the DataExtractor class with additional methods for qualifying-specific functionality

```python
from F1Archive.data.qualy_extractor_r1 import QualyExtractor
```

## Create a QualyExtractor object and get urls for required seasons
## get_race_urls method accepts int or str values
## stores a dict in the QualyExtractor object named year_urls 

## race urls are extracted with the DataExtractor class - these are required for extracting qualifying data for each race

```python
YEAR = '1998'

QDX = QualyExtractor()
QDX.get_race_urls(YEAR) # uses super class to get race-results urls

print(f"Race urls: {QDX.year_urls}")
```

## Get qualifying urls first, then get the corresponding results

```python
# use above urls to search for qualifying results on webpage with 'qualifying-0'
QDX.get_qualy_urls() 

QDX.year_qualy_results()
```

## Qualifying results stored in a dataframe with 'Position' and 'Time' attributes for each race
## 'Team-mate' column instantiated for later comparisons

```python
qdf_1998 = QDX.qualy_results[YEAR]
qdf_1998.head()
```

# Team-mate performance differences
## We want to fill in this column with a relative time to each driver's team-mate - this will be a plus or minus floating point value.

# The qualy_differences function is used to add team-mate comparisons to the dataframe

```python
from F1Archive.data_transforms.transformations1 import qualy_differences
import numpy as np
```

## first we want to get a list of the race names for the season we're looking at as the 'qualy_differences()' function takes them as an input.

```python
race_urls = QDX.qualy_urls[YEAR]
race_names = []
for race in race_urls:
    race_names.append(race.split('/')[6])
print(f"Races: {race_names}")
```

# this function uses the time_diff function to convert strings extracted from the website to datetime objects and adds to the 'team-mate' colums a floating-point number corresponding to difference to team-mate in seconds

```python
qdf_1998.replace(0, np.nan, inplace=True)
qdiff_df = qualy_differences(qdf_1998.copy(), race_names) # use .copy() to avoid inadvertently changing the original dataframe

qdiff_df.head()
```

## The dataframe above shows that comparisons between team-mates qualifying has been added where possible

# The qualy_relative_2_mean function is used to calculate each driver's qualifying time relative to the mean for all drivers

```python
from F1Archive.data_transforms.transformations1 import qualy_relative_2_mean
```

# Here, we pass a copy of the most recent dataframe and the race_names list
## The function ensures times are datetime objects and uses the avg_datetime function to calculate the field mean for a race 

## You can see the 'to_mean' columns addes to the dataframe. Negative values are those times faster than the mean.

```python
qr_df, _ = qualy_relative_2_mean(qdiff_df.copy(), race_names)
qr_df.head()
```

# The above times relative to mean can be used to visualize 'field-spread' for each race
## QualyDataExtractor has a method named 'stack_qualy_results' which facilitates putting data such as races into single columns

```python
"""
First we need to stack the dataframe so that position, race and time to mean are all contained in
single columns.
"""
stacked_df = QDX.stack_qualy_results(qr_df, race_names)
stacked_df.tail(15)
```

# Plotting field spread indicates more distance between the fastest drivers/cars and clustering in the mid-field
# It is relatively consistent throughout the season but tracks like Canada are particularly condensed

```python
from F1Archive.visualization.viz import plot_field_spread
```

```python
ax = plot_field_spread(stacked_df, race_names, year='1998')
```
